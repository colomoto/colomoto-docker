```python
from optboolnet.launch import control_fixpoint
import biolqm
```

```python
lqm = biolqm.load("http://ginsim.org/sites/default/files/Bladder_Model.zginml")
bn = biolqm.to_minibn(lqm)

inputs = {
    "GrowthInhibitors":1,
    "EGFR_stimulus":1,
    "FGFR3_stimulus":1
}
target = {"Apoptosis_b1": 1, "RB1": 1}
s = control_fixpoint(bn, max_size=2, target=target, inputs=inputs, exclude=["Apoptosis_b1","RB1"])
```
